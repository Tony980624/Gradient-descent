```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =  np.arange(-10,10,0.1)
y_real = x**2 + 2*x+8 +np.random.normal(0,1,size=x.shape)
plt.plot(x,y_real)
```

```python
def gradient_descent(initial_theta,learning_rate,iterations):
    a=  initial_theta[0]
    b = initial_theta[1]
    c= initial_theta[2]
    list = []
    losses = []
    for i in range(iterations):
        y_pred = a*x**2+b*x+c
        MSE = np.mean((y_pred - y_real)**2)
        losses.append(MSE)
        error = y_real-y_pred
        gradient_a = -2*np.mean(error*x**2)
        gradient_b = -2*np.mean(error*x)
        gradient_c = -2*np.mean(error)
        a = a -learning_rate*gradient_a
        b = b- learning_rate*gradient_b
        c = c-learning_rate*gradient_c
        list.append([a,b,c])
    return list,losses
    
```

```python
parameters,loss = gradient_descent([0,0,0],0.0002,30000)
plt.plot(loss[:100])
```

```python
parameters = np.array(parameters)
plt.plot(parameters[:,0],label='Para a')
plt.plot(parameters[:,1],label='Para b')
plt.plot(parameters[:,2],label='Para c')
plt.title("Parameters Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Parameter Value")
plt.legend()  # 

# 
plt.show()
```
